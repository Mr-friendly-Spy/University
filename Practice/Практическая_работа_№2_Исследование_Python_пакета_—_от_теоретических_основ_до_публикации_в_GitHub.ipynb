## **Описание работы**

Вы будете работать с уже созданным пакетом `calculator`, который имеет следующую структуру:

    calculator/
    ├── __init__.py
    ├── basic/
    │   ├── __init__.py
    │   ├── addition.py
    │   └── subtraction.py
    └── advanced/
        ├── __init__.py
        ├── exponentiation.py
        └── root.py

**Содержимое файлов:**

**`calculator/__init__.py`**

```python
# __init__.py in calculator
__all__ = ["basic", "advanced"]
```

**`calculator/basic/__init__.py`**

```python
# __init__.py in basic
from .addition import add
from .subtraction import subtract
```

**`calculator/basic/addition.py`**

```python
# addition.py
def add(a, b):
    return a + b
```

**`calculator/basic/subtraction.py`**

```python
# subtraction.py
def subtract(a, b):
    return a - b
```

**`calculator/advanced/__init__.py`**

```python
# __init__.py in advanced
from .exponentiation import power
from .root import square_root
```

**`calculator/advanced/exponentiation.py`**

```python
# exponentiation.py
def power(a, b):
    return a ** b
```

**`calculator/advanced/root.py`**

```python
# root.py
def square_root(a):
    return a ** 0.5
```

**Пример использования пакета:**

```python
# main.py
from calculator.basic import add, subtract
from calculator.advanced import power, square_root

print(add(2, 3))             # Вывод: 5
print(subtract(5, 2))        # Вывод: 3
print(power(2, 3))           # Вывод: 8
print(square_root(16))       # Вывод: 4.0
```

<font color='green'>Ваш ответ (Пакет продолжает быть работоспособным. Разница между абсолютным и относительным импортом в Python заключается в том, что абсолютный импорт указывает полный путь к модулю от корня проекта, а относительный — путь относительно текущего модуля. Оба подхода имеют свои преимущества и недостатки.


$Абсолютный импорт

Преимущества:
Ясность и читаемость — абсолютный путь чётко указывает на местоположение импортируемого модуля. Это упрощает понимание структуры проекта и поиск файлов, особенно в больших кодовых базах.
Стабильность — путь импорта не зависит от текущего местоположения импортирующего файла, что делает код более устойчивым к рефакторингу и перемещению файлов внутри проекта.
Универсальность — модули, использующие абсолютный импорт, могут быть легко запущены как самостоятельные скрипты без возникновения ошибок, связанных с контекстом пакета.

Недостатки:
В некоторых случаях абсолютный импорт становится длинным из-за подпапок или модулей, что может расширить инструкцию импорта в две строки.


$Относительный импорт

Преимущества:
Краткость — в глубоко вложенных пакетах относительный импорт может сделать строки импорта значительно короче, так как не требуется указывать полный путь от корня проекта.
Переносимость внутри пакета — если планируется переместить целый пакет в другое место в рамках более крупного проекта, относительные импорты внутри него останутся работоспособными, так как они ссылаются на другие модули относительно своего положения.
Избегание конфликтов — в некоторых редких случаях относительный импорт может помочь избежать конфликтов имён с модулями из стандартной библиотеки или сторонних пакетов, если есть модуль с таким же именем.

Недостатки:
Только внутри пакетов — модуль, использующий относительный импорт, не может быть запущен напрямую как основной скрипт, поскольку ему требуется контекст пакета для разрешения относительных путей.)</font>
